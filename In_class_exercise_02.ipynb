Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python


import time
import pandas as pd
import requests
from bs4 import BeautifulSoup


def scrape_backmarket(url,no):
    name = []
    spec = []
    war = []
    discount_price = []
    actual_price = []
    for i in range(1,no):
        response = requests.get(url +str(i))
        soup = BeautifulSoup(response.content, 'html.parser')
        data = soup.find_all('div', {'class': 'flex flex-col md:flex-1 md:justify-end'})

        for x in data:
            try:
                name.append(x.find('h2').get_text())
            except:
                name.append('na')

            try:
                spec.append(x.find('span', {'class': 'body-2-light duration-200 line-clamp-1 normal-case overflow-ellipsis overflow-hidden text-black transition-all'}).get_text().strip())
            except:
                spec.append('na')

            try:
                war.append(x.find('span', {'class': 'body-2-light text-black'}).get_text().strip())
            except:
                war.append('na')


            try:
                discount_price.append(x.find('span', {'class': 'body-2-bold text-black'}).get_text().strip())
            except:
                discount_price.append('na')

            try:
                actual_price.append(x.find('div', {'class': 'body-2-light text-primary-light line-through'}).get_text().strip())
            except:
                actual_price.append('na')

    df = pd.DataFrame({'Name': name, 'Type': spec, 'Discounted Price': discount_price,
                       'Actual Price': actual_price, 'Warranty': war})

    return df


url_page1 = 'https://www.backmarket.com/en-us/l/smartphones/0744fd27-8605-465d-8691-3b6dffda5969?page='
url_page2 = 'https://www.backmarket.com/en-us/l/iphone/e8724fea-197e-4815-85ce-21b8068020cc?page='


df_page1 = scrape_backmarket(url_page1,34)
df_page2 = scrape_backmarket(url_page2,5)


combined_df = pd.concat([df_page1, df_page2], ignore_index=True)

df1 = combined_df.replace('\n','', regex=True)
df1


```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):
import requests
import pandas as pd
from bs4 import BeautifulSoup


def collect_articles(query, num_articles):
    base_url = "https://scholar.google.com/scholar"
    collected_articles = []

    while len(collected_articles) < num_articles:
        params = {
            "q": query,
            "hl": "en",
            "start": len(collected_articles),
        }

        response = requests.get(base_url, params=params)
        soup = BeautifulSoup(response.text, "html.parser")


        entries = soup.find_all("div", class_="gs_ri")

        for entry in entries:
            article_data = {}


            title = entry.find("h3", class_="gs_rt")
            article_data["Title"] = title.text if title else "N/A"


            venue = entry.find("div", class_="gs_a")
            article_data["Venue"] = venue.text if venue else "N/A"


            year = entry.find("div", class_="gs_a").text.split(" - ")[-1]
            article_data["Year"] = year if year else "N/A"


            authors = entry.find("div", class_="gs_a").text.split(" - ")[0]
            article_data["Authors"] = authors if authors else "N/A"


            abstract = entry.find("div", class_="gs_rs")
            article_data["Abstract"] = abstract.text if abstract else "N/A"

            collected_articles.append(article_data)

    return collected_articles


query = "Data Science"
num_articles = 1000


articles = collect_articles(query, num_articles)


for i, article in enumerate(articles, start=1):
    print(f"Article {i}:")
    print(f"Title: {article['Title']}")
    print(f"Venue/Journal: {article['Venue']}")
    print(f"Year: {article['Year']}")
    print(f"Authors: {article['Authors']}")
    print(f"Abstract: {article['Abstract']}")
    print("\n")

```

Question 4 (10 points): Write python code to collect 1000 posts from Twitter, or Facebook, or Instagram. You can either use hashtags, keywords, user_name, user_id, or other information to collect the data.

The following information needs to be collected:

(1) User_name

(2) Posted time

(3) Text

```python
import instaloader

import pandas as pd

L = instaloader.Instaloader()

profile = instaloader.Profile.from_username(L.context, 'rvcjinsta')


collected_data = []


num_posts_to_collect = 1100


for post in profile.get_posts():
    if len(collected_data) >= num_posts_to_collect:
        break

    user_name = post.owner_username
    posted_time = post.date_utc
    text = post.caption

    collected_data.append({
        'User_name': user_name,
        'Posted_time': posted_time,
        'Text': text,
    })

df = pd.DataFrame(collected_data)

df
```
